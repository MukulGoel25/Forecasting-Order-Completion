FORECASTING FUTURE ORDER COMPLETION

Importing necessary library

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

LOADING THE DATA 

```python
data = pd.read_csv('Data.csv')
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
data.set_index('Date', inplace=True)
```

Plot the data

```python
plt.plot(data.index, data['Order Comp'], marker='o', linestyle='-', color='b')
#labels
plt.xlabel('Date')
plt.ylabel('Order Comp')
plt.title('Order Comp Over Time')
plt.xticks(rotation=45)
# Display the plot
plt.show()
```

From the plot is is evident that the data is not stationary
ADF Test to support the argument

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(data['Order Comp'])
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")

if result[1] > 0.05:
    print("The time series is non-stationary.")
else:
    print("The time series is stationary.")
```

Making the data stationary 

```python
data['Differenced'] = data['Order Comp'] - data['Order Comp'].shift(1)
data.dropna(inplace=True)
```

ADF Test for check stationarity with differencing

```python
result = adfuller(data['Differenced'])
print(f"Differenced ADF Statistic: {result[0]}")
print(f"Differenced p-value: {result[1]}")
```

```python
print (2.870891596238548e-09 > 0.05)
```

Single differencing Worked

Plotting the Stationary Data

```python
plt.plot(data['Differenced'])
plt.title('Differenced Time Series')
plt.xticks(rotation=45)
plt.show()
```

KS Test for Strict Stationarity

```python
from scipy.stats import kstest

statistic, p_value = kstest(data['Differenced'], 'norm', args=(data['Differenced'].mean(), data['Differenced'].std()))

print(f"KS Statistic: {statistic}")
print(f"p-value: {p_value}")
```

High KS test p value suggests that the data is following strict stationary

determining p and q for ARIMA model by plotting ACF and PCF

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
y = data['Order Comp']
plot_acf(y) 
plot_pacf(y)
plt.show()
```

ACF and PACF plot shows the value of p and q to be 2 or 3 

```python
train_size = int(len(data) * 0.90)  # Use 90% for training
train_data = data['Order Comp'][:train_size]
test_data = data['Order Comp'][train_size:]
# Fit ARIMA model on training data
model = ARIMA(train_data, order=(2, 1, 2))
fitted_model = model.fit()
# Forecast values for the test period
forecast = fitted_model.forecast(steps=len(test_data))
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Compute metrics
mae = mean_absolute_error(test_data, forecast)
mse = mean_squared_error(test_data, forecast)
rmse = mse**0.5
mape = (abs((test_data - forecast) / test_data).mean()) * 100

# Print results
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAPE: {mape:.2f}%")
```

```python
model = ARIMA(data['Differenced'], order=(2, 1, 2))
fitted_model = model.fit()

# Forecasting future values
forecast_steps = 30
forecast = fitted_model.forecast(steps=forecast_steps)

# Displaying forecasted values
print("Forecasted values:")
print(forecast)
last_observed = data['Order Comp'].iloc[-1]

# Converting differenced forecast back to original scale
original_forecast = forecast.cumsum() + last_observed

# Displaying original scale forecast
print("Forecasted values (original scale):")
print(original_forecast)
plt.figure(figsize=(12, 6))
plt.plot(data['Order Comp'], label='Observed', color='blue')
plt.plot(original_forecast, label='Forecast', color='red')
plt.title('Order Completion Forecast')
plt.xlabel('Date')
plt.ylabel('Order Comp')
plt.legend()
plt.grid()
plt.show()
```
